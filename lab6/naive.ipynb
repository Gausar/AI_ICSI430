```python
import os

test_ham_path = "C:\\Users\\Gausar\\OneDrive\\Documents\\2024namar\\AI\\lab6\\spam_data\\dev\\ham"

test_ham_list = os.listdir(test_ham_path)
ham_tests = []
for file_name in test_ham_list:
    file_path = os.path.join(test_ham_path, file_name)
    if os.path.isfile(file_path):
        with open(file_path, 'r', encoding='utf-8') as file:
            content = file.read()
            ham_tests.append((content, 0))
print(ham_tests)

```

```python
from collections import defaultdict
import numpy as np

def build_vocabulary(emails):
    vocabulary = set()
    word_counts = defaultdict(lambda: [0, 0])  # 0: non-spam, 1: spam
    
    for email, label in emails:
        words = tokenize(email)
        for word in words:
            vocabulary.add(word)
            word_counts[word][label] += 1
            
    return vocabulary, word_counts

def train_naive_bayes(emails, vocabulary, word_counts):
    total_emails = len(emails)
    spam_count = sum(1 for _, label in emails if label == 1)
    non_spam_count = total_emails - spam_count

    # spam bolon non-spam emailiin magadlal 
    prob_spam = spam_count / total_emails
    prob_non_spam = non_spam_count / total_emails

    # ug buriin taarah magadlal bodogdoj bna
    word_prob_spam = {}
    word_prob_non_spam = {}
    for word in vocabulary:
        word_prob_spam[word] = (word_counts[word][1] + 1) / (spam_count + len(vocabulary))
        word_prob_non_spam[word] = (word_counts[word][0] + 1) / (non_spam_count + len(vocabulary))

    return prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam


def tokenize(email):
    return email.lower().split() 

def predict_naive_bayes(email, prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam):
    words = tokenize(email)
    
    # het baga too overflow uusgeh tul log avch urjihiin orond nemsen
    spam_score = np.log(prob_spam)
    non_spam_score = np.log(prob_non_spam)
    
    for word in words:
        if word in word_prob_spam:
            spam_score += np.log(word_prob_spam[word])
        if word in word_prob_non_spam:
            non_spam_score += np.log(word_prob_non_spam[word])
            
    return 1 if spam_score > non_spam_score else 0
```

```python
def compute_confusion_matrix(test_emails, prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam):
    TP = FP = TN = FN = 0
    
    for email, actual_label in test_emails:
        predicted_label = predict_naive_bayes(email, prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam)
        
        if predicted_label == 1 and actual_label == 1:
            TP += 1
        elif predicted_label == 1 and actual_label == 0:
            FP += 1
        elif predicted_label == 0 and actual_label == 0:
            TN += 1
        elif predicted_label == 0 and actual_label == 1:
            FN += 1

    return TP, FP, TN, FN


train_emails = spam_files + ham_files
test_emails = spam_tests + ham_tests

vocabulary, word_counts = build_vocabulary(train_emails)
prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam = train_naive_bayes(train_emails, vocabulary, word_counts)
TP, FP, TN, FN = compute_confusion_matrix(test_emails, prob_spam, prob_non_spam, word_prob_spam, word_prob_non_spam)

print("Confusion Matrix:")
print(f"True Positive (TP): {TP}")
print(f"False Positive (FP): {FP}")
print(f"True Negative (TN): {TN}")
print(f"False Negative (FN): {FN}")
```

```python

precision = TP / (TP + FP)
recall = TP / (TP + FN)
accuracy = (TP + TN) / (TP + TN + FP + FN)
f1_score = 2 * (precision * recall) / (precision + recall)


print("Unelgee:")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"Accuracy: {accuracy:.2f}")
print(f"F1 Score: {f1_score:.2f}")
```
