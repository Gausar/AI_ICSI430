```python
def hill_climbing(G, init_tour):
    current_tour = init_tour[:]
    current_distance = calculate_total_distance(current_tour, G)

    while True:
        neighbors = []

        for i in range(1, len(current_tour) - 1):
            for j in range(i + 1, len(current_tour)):
                neighbor = current_tour[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i] 
                neighbors.append(neighbor)

        best_neighbor = None
        best_distance = current_distance

        for neighbor in neighbors:
            distance = calculate_total_distance(neighbor, G)
            if distance < best_distance:
                best_distance = distance
                best_neighbor = neighbor

        if best_neighbor is None:
            break

        current_tour = best_neighbor
        current_distance = best_distance

    return current_tour, current_distance
```

```python
import random
import math

def calculate_total_distance(route, data):
    total_distance = 0
    for i in range(len(route) - 1):
        city_a = route[i]
        city_b = route[i + 1]
        for edge in data['edges']:
            if (city_a, city_b) == (edge[0], edge[1]) or (city_b, city_a) == (edge[0], edge[1]):
                total_distance += edge[2]['weight']
                break
    city_a = route[-1]
    city_b = route[0]
    for edge in data['edges']:
        if (city_a, city_b) == (edge[0], edge[1]) or (city_b, city_a) == (edge[0], edge[1]):
            total_distance += edge[2]['weight']
            break
    return total_distance

# Hill-Climbing Algorithm
def hill_climbing(data):
    cities = data['nodes']

    current_solution = random.sample(cities, len(cities))
    current_distance = calculate_total_distance(current_solution, data)

    improving = True
    while improving:
        improving = False
        for i in range(len(cities)):
            for j in range(i + 1, len(cities)):
                neighbor = current_solution[:]
                neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
                neighbor_distance = calculate_total_distance(neighbor, data)
                if neighbor_distance < current_distance:
                    current_solution = neighbor
                    current_distance = neighbor_distance
                    improving = True
                    break
            if improving:
                break

    return current_solution, current_distance

#Simulated annealing algorithm
def simulated_annealing(data, initial_temp=1000, cooling_rate=0.99, min_temp=1):
    cities = data['nodes']
    current_solution = random.sample(cities, len(cities))
    current_distance = calculate_total_distance(current_solution, data)

    temperature = initial_temp

    while temperature > min_temp:
        i, j = random.sample(range(len(cities)), 2)
        neighbor = current_solution[:]
        neighbor[i], neighbor[j] = neighbor[j], neighbor[i]
        neighbor_distance = calculate_total_distance(neighbor, data)

        if neighbor_distance < current_distance or random.uniform(0, 1) < math.exp((current_distance - neighbor_distance) / temperature):
            current_solution = neighbor
            current_distance = neighbor_distance

        temperature *= cooling_rate

    return current_solution, current_distance

hill_climb_solution, hill_climb_distance = hill_climbing(data)
sa_solution, sa_distance = simulated_annealing(data)

print("Hill-Climbing Solution:", hill_climb_solution)
print("Hill-Climbing Distance:", hill_climb_distance)
print("Simulated Annealing Solution:", sa_solution)
print("Simulated Annealing Distance:", sa_distance)
```
